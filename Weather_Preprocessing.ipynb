```python
import pandas as pd

# Load your dataset (replace 'your_dataset.csv' with your file name)
file_path = '/content/gdrive/MyDrive/swedish_cities_environmental.csv'
data = pd.read_csv(file_path, encoding='utf-8')

print(data['city'].unique())

```

```python
Skelleftea_data = data[data['city'] == 'Skelleftea']

# Save the filtered data to a new file
output_path = 'Skelleftea_only_data.csv'
Skelleftea_data.to_csv(output_path, index=False)

print(f"Data for Skelleftea has been saved to {output_path}")
```

```python
import pandas as pd

# Load your dataset
file_path = '/content/Skelleftea_only_data.csv'  # Replace with the actual dataset file name
data = pd.read_csv(file_path)

# Convert the 'timestamp' column to datetime format
data['timestamp'] = pd.to_datetime(data['timestamp'])

# Extract the date (without the time) and create a new column for grouping
data['date'] = data['timestamp'].dt.date

# Group by the 'date' column and aggregate the data (e.g., mean, sum, etc.)
Skelleftea_data = data.groupby('date').agg({
    'temperature': 'mean',               # Average temperature
    'humidity': 'mean',                  # Average humidity
    'pressure': 'mean',                  # Average pressure
    'weather_description': 'first',      # Take the first weather description of the day
})

# Save the aggregated data to a new file
output_path = 'Skelleftea_data.csv'
Skelleftea_data.to_csv(output_path)

print(f"Aggregated data has been saved to {output_path}")
```
